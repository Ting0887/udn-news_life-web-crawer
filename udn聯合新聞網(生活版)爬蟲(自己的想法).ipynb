```python
import requests
import json
from bs4 import BeautifulSoup
#生活新聞
news = []
for i in range(0,7):
    if i==0:
        url = 'https://udn.com/api/more?&channelId=2&type=subcate_articles&cate_id=6649&sub_id=7266&totalRecNo=62&is_paywall=0&is_bauban=0&is_vision=0'
    else:    
        url = 'https://udn.com/api/more?page={}&channelId=2&type=subcate_articles&cate_id=6649&sub_id=7266&totalRecNo=64&is_paywall=0&is_bauban=0&is_vision=0'.format(i)
    html = requests.get(url)
    response = json.loads(html.text)
    #print(response['lists'])
    for item in response['lists']:
        link = 'https://udn.com/'+item['titleLink'] #連結
        title = item['title'] #標題
        content = item['paragraph'] #內容
        time = item['time'] #時間
        date = time['date'] #時間+日期
        
        news.append({'新聞標題':title,
                     '連結':link,
                     '內容':content,
                     '報導日期':date})
        
file = '生活新聞.json'
with open(file,'w',encoding='utf-8') as data:
    json.dump(news,data,ensure_ascii=False,indent=2)
data.close()
```
